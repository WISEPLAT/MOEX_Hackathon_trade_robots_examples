```python
# Сбор данных по торгам в цикле
import pandas as pd
from datetime import datetime, timedelta
df = pd.DataFrame([])
#Устанавливаем начальную дату
start_date = datetime.strptime('2022-01-01', '%Y-%m-%d')

#Устанавливаем конечную дату
end_date = start_date + timedelta(days=699)

while start_date <= end_date:
# Извлекаем данные за текущую дату
    data = stocks.tradestats(date=start_date.strftime('%Y-%m-%d'))
    df = df.append(data, ignore_index=True) 
# Здесь обрабатываем извлеченные данные...
    print(start_date,' всего записей: ',len(data))
# Переходим к следующей дате
    start_date += timedelta(days=1)
```
