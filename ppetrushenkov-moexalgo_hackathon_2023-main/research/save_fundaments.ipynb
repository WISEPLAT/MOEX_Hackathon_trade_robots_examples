```python
from tqdm import tqdm
```

```python
tickers = []
sectors = []
capitalizations = []

pbar = tqdm(os.listdir('data/d1/'))
for ticker_name in pbar:
    try:
        ticker_name = ticker_name.split('.')[0]
        pbar.set_description(ticker_name)
        tickers.append(ticker_name)

        ticker = yf.Ticker(f'{ticker_name}.ME')
        sector = ticker.info.get('sector')
        if sector is not None:
            sectors.append(sector)
        else:
            sectors.append('-')
        
        market_cap = ticker.info.get('marketCap')
        if market_cap is not None:
            capitalizations.append(market_cap)
        else:
            capitalizations.append('-')
    
    except Exception as e:
        print(ticker_name)
        print(e)


ticker_info = {
    'Ticker': tickers,
    'Sector': sectors,
    'MarketCap': capitalizations
}
```

```python
ticker_info = pd.DataFrame(ticker_info)
```

```python
ticker_info.to_csv('data/ticker_info.csv', index=False)
```
