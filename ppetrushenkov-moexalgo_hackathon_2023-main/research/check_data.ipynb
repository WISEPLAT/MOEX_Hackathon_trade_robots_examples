```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover, resample_apply
```

```python
import pandas_ta as pta
```

```python
sber = data[['begin', 'open', 'high', 'low', 'close', 'volume']]
```

```python
sber.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
sber.set_index('Date', inplace=True)
```

```python
sber.head()
```

```python
def ma(price, period=21):
    return pta.sma(price, period).to_numpy()

def supertrend(data):
    st = pta.supertrend(data.High.s, data.Low.s, data.Close.s, length=21, multiplier=3)
    return st.iloc[:, 0].to_numpy()
```

```python
class Supertrend(Strategy):
    def init(self):
        self.ma_short = self.I(supertrend, self.data)
        self.ma_long = resample_apply(rule='1M', func=ma, series=self.data.Close.s)

    def next(self):
        if self.data.Close.s[-1] > self.ma_long[-1] and crossover(self.data.Close.s, self.ma_short):
            self.buy()
        elif self.data.Close.s[-1] < self.ma_long[-1] and crossover(self.ma_short, self.data.Close.s):
            self.sell()
```

```python
def sma(price, period=21):
    return pta.sma(price)
```

```python
bt = Backtest(sber, Supertrend, commission=.002,
              exclusive_orders=True)
stats = bt.run()
print(stats)
bt.plot()
```

```python
def save_all_data(tickers: list, date_from: str = '2010-01-01', date_to: str = 'today', period='D'):
    for ticker in tickers:
        try:
            data = Ticker(ticker)
            candles = data.candles(date=date_from, till_date=date_to, period=period)
            candles.to_csv(f'data/d1/{ticker}.csv')
        except Exception as e:
            print(ticker)
```

```python
class Supertrend(Strategy):
    def init(self):
        self.ma_short = self.I(supertrend, self.data)
        self.ma_long = resample_apply(rule='1M', func=ma, series=self.data.Close.s)

    def next(self):
        if self.data.Close.s[-1] > self.ma_long[-1] and crossover(self.data.Close.s, self.ma_short):
            self.buy()
        elif self.data.Close.s[-1] < self.ma_long[-1] and crossover(self.ma_short, self.data.Close.s):
            self.sell()
```

```python
def ma(price, period=21):
    return ta.SMA(price, period).to_numpy()
```

```python
def SMA(series, n):
    from backtesting.test import SMA
    return SMA(series, n)
```

```python
class MaTrendTrading(Strategy):
    def init(self):
        self.long_ma = self.I(SMA, self.data.Close.s, n=252)
        self.short_ma = self.I(SMA, self.data.Close.s, n=21)
    
    def next(self):
        if crossover(self.short_ma, self.long_ma):
            self.buy()
        elif crossover(self.long_ma, self.short_ma):
            self.sell()
```

```python
bt = Backtest(sber, MaTrendTrading, commission=.002, exclusive_orders=True)
stats = bt.run()
print(stats)
```
