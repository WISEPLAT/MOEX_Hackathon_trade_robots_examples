```python
from moexalgo import Market, Ticker

import pandas as pd
import numpy as np

from datetime import datetime
```

```python
tradestats = pd.DataFrame()
obstats = pd.DataFrame()
orderbook = pd.DataFrame()

for ticker_name in ['AFLT', 'BANE', 'DSKY', 'FEES', 'GAZP', 'PHOR', 'POSI', 'SBER', 'YNDX', 'MTLR']:
    ticker = Ticker(ticker_name)
    
    t = ticker.tradestats(date='2023-11-01', till_date='2023-11-30')
    t.drop(columns='systime', inplace=True)
    tradestats = pd.concat([tradestats, t])

    obs = ticker.obstats(date='2023-11-01', till_date='2023-11-30')
    obs.drop(columns='systime', inplace=True)
    obstats = pd.concat([obstats, obs])
    
    ord = ticker.orderstats(date='2023-11-01', till_date='2023-11-30')
    ord.drop(columns=['systime'], inplace=True)
    orderbook = pd.concat([orderbook, ord])
```

```python
tradestats.to_csv('backtest_tradestats.csv')
obstats.to_csv('backtest_obstats.csv')
orderbook.to_csv('backtest_orderbook.csv')
```
