```python
import pandas as pd
import json
```

```python
data = pd.read_excel("ListingSecurityList.xlsx")
```

```python
data_short = data[['TRADE_CODE', 'INSTRUMENT_TYPE', 'EMITENT_FULL_NAME']]
```

```python
data_drop_nan = data_short.dropna()
```

```python
data_drop_nan['INSTRUMENT_TYPE'].unique()
```

```python
instruments = ['Акция обыкновенная', 'Депозитарные расписки иностранного эмитента на акции',
       'Акции иностранного эмитента', 'Акция привилегированная']
```

```python
result_data = data_drop_nan.loc[data_drop_nan['INSTRUMENT_TYPE'].isin(instruments)]
```

```python
result_data.head()
```

```python
result_dict = {}
for index, row in result_data.iterrows():
    result_dict[row['TRADE_CODE']] = {'INSTRUMENT_TYPE': row['INSTRUMENT_TYPE'], 
                                      'EMITENT_FULL_NAME': row['EMITENT_FULL_NAME']}
```

```python
result_dict
```

```python
with open('trade_data.json', 'w', encoding='utf-8') as outfile:
    json.dump(result_dict, outfile)
```

```python
with open('trade_data.json', 'r', encoding='utf-8') as file:
    result_dict_read = json.load(file)
```
