```python
import pandas as pd
import time
import numpy as np
from tqdm import tqdm
from datetime import datetime
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df21_all = pd.read_csv('tradestats_2021.csv', sep=';')
df21 = df21_all[df21_all['secid'].isin(["GAZP"])]
display(df21)
del df21_all
```

```python
df20_all = pd.read_csv('tradestats_2020.csv', sep=';')
df20 = df20_all[df20_all['secid'].isin(["GAZP"])]
display(df20)
del df20_all
```

```python
df22_all = pd.read_csv('tradestats_2022.csv', sep=';')
df22 = df22_all[df22_all['secid'].isin(["GAZP"])]
display(df22)
del df22_all
```

```python
df23_all = pd.read_csv('tradestats_2023.csv', sep=';')
df23 = df23_all[df23_all['secid'].isin(["GAZP"])]
display(df23)
del df23_all
```

```python
df = pd.concat([df20, df21, df22, df23], axis=0, ignore_index=True)
display(df)
```

```python
df = df.drop(columns = ['SYSTIME', 'secid'])
```

```python
df
```

```python
df['date'] = df['tradedate'] +" "+ df['tradetime']
df['date'] = pd.to_datetime(df['date'])
df.drop(columns = ['tradedate','tradetime'], inplace = True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
def getWeekday(date):
  return date.weekday()
```

```python
df['weekday']=df['date'].apply(getWeekday)
```

```python
df
```

```python
encoder = OneHotEncoder(drop='first', categories='auto',sparse=False)
weekday_values = df['weekday'].values.reshape(-1, 1)
one_hot_encoded = encoder.fit_transform(weekday_values)
one_hot_df = pd.DataFrame(one_hot_encoded, columns=[f'weekday_{i}' for i in range(1, 5)])
df = pd.concat([df, one_hot_df], axis=1).drop('weekday', axis=1)
```

```python
df.info()
```

```python
df
```

```python
df.drop(columns=['date'], inplace = True)
```

```python
df['target']=df['pr_low'].shift(-1)
df.dropna(inplace=True)
```

```python
df
```

```python
target = df['target']
df.drop(columns=['target'], inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.datasets import load_wine
import optuna
from optuna.samplers import TPESampler
from catboost import CatBoostRegressor
import pickle
from sklearn.metrics import mean_squared_error
```

```python
def objective(trial):
    param = {}
    param['learning_rate'] = trial.suggest_discrete_uniform("learning_rate", 0.001, 0.02, 0.001)
    param['depth'] = trial.suggest_int('depth', 9, 15)
    param['l2_leaf_reg'] = trial.suggest_discrete_uniform('l2_leaf_reg', 1.0, 5.5, 0.5)
    param['min_child_samples'] = trial.suggest_categorical('min_child_samples', [1, 4, 8, 16, 32])
    param['grow_policy'] = 'Depthwise'
    param['iterations'] = 100
    param['eval_metric'] = 'RMSE'
    param['od_wait'] = 20
    param['random_state'] = 42
    param['logging_level'] = 'Silent'

    regressor = CatBoostRegressor(**param)

    regressor.fit(X_train.copy(), y_train.copy())
    loss = mean_squared_error(y_test, regressor.predict(X_test.copy()))
    return loss
```

```python
optuna.logging.set_verbosity(optuna.logging.WARNING)

sampler = TPESampler(seed=1)
study = optuna.create_study(study_name="catboost", direction="maximize", sampler=sampler)
study.optimize(objective, n_trials=10)
```

```python
print("Number of finished trials: ", len(study.trials))
print("Best trial:")
trial = study.best_trial
print("  Value: ", trial.value)
print("  Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))
```
