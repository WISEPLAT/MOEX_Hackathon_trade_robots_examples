```python
def process_df(df):
    df['Дата'] = pd.to_datetime(df['Дата'])
    df['Время'] = pd.to_datetime(df['Время'], format='%H:%M:%S').dt.time
    df['Дата+Время'] = pd.to_datetime(df['Дата'].astype(str) + ' ' + df['Время'].astype(str))
    df.drop(columns=['Дата', 'Время'], inplace=True)
    df.rename(columns={'Дата+Время': 'Время'}, inplace=True)
    df['Время'] = df['Время'].dt.to_pydatetime()
    return df

# Применяем функцию к каждому DataFrame
df1_processed = process_df(orders)
df2_processed = process_df(trades)
df3_processed = process_df(books)
```

```python
result = pd.merge(df1_processed, df2_processed, on='Время', how='outer')
ticker_df = pd.merge(result, df3_processed, on='Время', how='outer')
ticker_df = ticker_df.dropna()
```

```python
algo_df_test = ticker_df.copy(deep=True)
algo_df_test
```

```python
initial_balance = 100000000  # Начальный баланс
take_profit_multiplier = 1.1  # Take profit на 2% выше цены покупки
stop_loss_multiplier = 0.90  # Stop loss на 2% ниже цены покупки
shares_to_buy = 100  # Количество акций для покупки

# Расчет скользящего среднего давления покупок и продаж
window_size = 3
algo_df_test['SMA_pressure_buy'] = algo_df_test['Кол-во поставленных заявок (покупка)'].rolling(window=window_size).mean()
algo_df_test['SMA_pressure_sell'] = algo_df_test['Кол-во поставленных заявок (продажа)'].rolling(window=window_size).mean()

# Определение пороговых значений
threshold_buy = algo_df_test['SMA_pressure_buy'].quantile(0.75)
threshold_sell = algo_df_test['SMA_pressure_sell'].quantile(0.75)

# Расчет изменения давления
algo_df_test['pressure_buy_growth'] = algo_df_test['SMA_pressure_buy'].diff()
algo_df_test['pressure_sell_growth'] = algo_df_test['SMA_pressure_sell'].diff()

# Торговая функция с учетом take_profit и stop_loss
def trading_decision(row, balance, shares_to_buy):
    # Цены для расчета take_profit и stop_loss
    entry_price = row['Последняя цена за период']
    take_profit_price = entry_price * take_profit_multiplier
    stop_loss_price = entry_price * stop_loss_multiplier

    if row['SMA_pressure_buy'] > threshold_buy and row['pressure_buy_growth'] < 0:
        # Проверка достаточности баланса для покупки
        cost_of_purchase = entry_price * shares_to_buy
        if balance >= cost_of_purchase:
            balance -= cost_of_purchase
            return 'long', balance, take_profit_price, stop_loss_price
        else:
            return 'hold', balance, None, None
    elif row['SMA_pressure_sell'] > threshold_sell and row['pressure_sell_growth'] < 0:
        # Продажа возможна только если есть акции в портфеле (здесь не рассматривается)
        return 'short', balance, None, None
    else:
        return 'hold', balance, None, None
```

```python
balance = initial_balance
take_profit_price = None
stop_loss_price = None

# Применение торговой функции к каждой строке DataFrame
for index, row in algo_df_test.iterrows():
    decision, balance, take_profit_price, stop_loss_price = trading_decision(row, balance, shares_to_buy)
    algo_df_test.at[index, 'trading_decision'] = decision
    
    # Проверка условий для закрытия позиции по take_profit или stop_loss
    if decision == 'long' and (row['Последняя цена за период'] >= take_profit_price or row['Последняя цена за период'] <= stop_loss_price):
        # Расчет прибыли или убытка и обновление баланса
        profit_or_loss_per_share = row['Последняя цена за период'] - (balance / shares_to_buy)
        total_profit_or_loss = profit_or_loss_per_share * shares_to_buy
        balance += total_profit_or_loss
        # Сброс take_profit и stop_loss после закрытия позиции
        take_profit_price = None
        stop_loss_price = None

# Вывод результатов торговли и конечного баланса
print(algo_df_test[['Время', 'trading_decision']])
print(f"Final balance: {balance}")
```
