# Рабочая версия скачивания данных для алгоритма

```python
output_df = pd.DataFrame(columns=['TRADEDATE', 'Name', 'Open', 'Close', 'Diff'])

time_i = start_time
while time_i != end_time:
    
    try:
        data_df = (
            Ticker('YNDX')
            .tradestats(date=str((time_i - timedelta(days=1)).date()), till_date=str(time_i.date()))
        )

        temp_df = (
            data_df.iloc[[0, stock_df.shape[0] - 1]][['tradedate', 'ticker', 'pr_open', 'pr_close']].head(1))
        temp_df['pr_open'] = (
            data_df.iloc[[0, stock_df.shape[0] - 1]][['tradedate', 'ticker', 'pr_open', 'pr_close']]
            .head(1)['pr_open'].to_list()[0])
        temp_df['pr_close'] = (
            data_df.iloc[[0, stock_df.shape[0] - 1]][['tradedate', 'ticker', 'pr_open', 'pr_close']]
            .tail(1)['pr_close'].to_list()[0])

        stock_df = temp_df.rename({'tradedate': 'TRADEDATE', 'ticker': 'Name', 'pr_open':'Open', 'pr_close': 'Close'}, axis=1)
        stock_df['Diff'] = stock_df['Close'] - stock_df['Open']

        output_df = pd.concat([output_df, stock_df], ignore_index=True)
        
    except:
        print('Something went wrong for', str((time_i - timedelta(days=1)).date()), str(time_i.date()))
        
    time_i = time_i + timedelta(days=1)
    
# output_name = os.path.join(dirpath, stock + '.csv')
output_name = 'YNDX.csv'
stock_df.to_csv(output_name)
```
